```python
#libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
#preprocessing
def preprocess(text):

  #lowercasing
  text = text.lower()

  #remove punctuation
  text = re.sub(r'[^a-zA-Z\s]','',text)

  #tokenization
  tokens = word_tokenize(text)

  #removing stopwords
  y = []
  for i in tokens:
    if i not in stop_words:
      y.append(i)

  #lammatizing
  x = []
  for i in y:
    x.append(lemmatizer.lemmatize(i))

  return " ".join(x)
```

```python
df['tags'] = df['tags'].apply(preprocess)
```

```python
#extracting indices with index
indices = pd.Series(df.index,index=df['title'])
indices.head()
```

#vectorization

```python
#tfidf
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=10000,ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

#Recommendation using cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#recomendation function
def recommend(title,n=10):
  if title not in indices:
    return ['Book not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df.iloc[similar_idx]
```

```python
recommend("Spider's Web")
```
